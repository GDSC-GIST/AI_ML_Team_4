### The "Hello World' of neural networks 
* Dense: define a layer connected
* Loss function: measures 'how good or bad the guess was?"
* Do not return exact value 


```python
from tensorflow import keras
import numpy as np 

model = keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
xs = np.array([-1,0,1,2,3,4,5,9], dtype = float)
ys = np.array([-3,-1,1,3,5,7,9, 17 ], dtype = float)

model.fit(xs, ys, epochs = 500)

print(model.predict([10]))
```

### Define and  Compile the Neural Network 
* Create simplest possible NN -> 1 layer with 1 neuron

* Specify two functions: 
    - loss function : measures the guessed answers against the known answers
    - optimizer : try to minimize the loss 
  

```python
model = keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
```

### Providing the Data
* using Numpy array

```python
xs = np.array([-1,0,1,2,3,4,5,9], dtype = float)
ys = np.array([-3,-1,1,3,5,7,9, 17 ], dtype = float)
```

### Training the Neural Network
* it learns relation between Xs and Ys in the **model.fit** call
* going through the loop of loss and optimizer

```python
model.fit(xs, ys , epochs = 500)
```

```python
print(model.predict([10.0]))
```

### W1_Assignment

```python
def house_model(y_new):
    xs = [0,1,2,3,4,5,6,9]
    ys = [5,10,15,20,25,30,35,50]
    model = keras.Sequential([keras.layers.Dense( units = 1, input_shape = [1])])
    model.compile(optimizer = 'sgd', loss = 'mean_squared_error')
    model.fit(xs, ys, epochs = 500)
    pred = model.predict(y_new)[0]
    return pred

pred = house_model([7.0])
print(pred)
    
```
